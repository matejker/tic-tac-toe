```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(0, 1.01, 0.01)

plt.figure(figsize=(10, 10))
plt.axis('off')

plt.plot([1, 2, 2, 1, 1], [1, 1, 2, 2, 1], 'k:', alpha=.2)

plt.plot([0.95, 2.05], [0.95, 2.05], 'g--', alpha=.7)
plt.plot([2.05, 0.95], [0.95, 2.05], 'g--', alpha=.7)
plt.plot([1.5, 1.5], [0.95, 2.05], 'g--', alpha=.7)
plt.plot([0.95, 2.05], [1.5, 1.5], 'g--', alpha=.7)

plt.plot(.8  * np.cos(x[:76] * 2 * np.pi) + 1.5, .8 * np.sin(x[:76] * 2 * np.pi) + 1.5, 'c')
plt.plot(.85  * np.cos(x[:51] * 2 * np.pi) + 1.5, .85 * np.sin(x[:51] * 2 * np.pi) + 1.5, 'c')
plt.plot(.9  * np.cos(x[:26] * 2 * np.pi) + 1.5, .9 * np.sin(x[:26] * 2 * np.pi) + 1.5, 'c')

plt.scatter(.8 * np.cos(x[75] * 2 * np.pi) + 1.5, .8 * np.sin(x[75] * 2 * np.pi) + 1.5, marker='>', color="c")
plt.scatter(.85 * np.cos(x[50] * 2 * np.pi) + 1.5, .85 * np.sin(x[50] * 2 * np.pi) + 1.5, marker='v', color="c")
plt.scatter(.9 * np.cos(x[25] * 2 * np.pi) + 1.5, .9 * np.sin(x[25] * 2 * np.pi) + 1.5, marker='<', color="c")

plt.text(1.5, 2.10, r"$t_y$", fontsize=20)
plt.text(2.10, 1.5, r"$t_x$", fontsize=20)
plt.text(2.12, 0.87, r"$t_{AC}$", fontsize=20)
plt.text(1, 0.87, r"$t_{BD}$", fontsize=20)

plt.text(.9 * np.cos(x[70] * 2 * np.pi) + 1.5, .9 * np.sin(x[70] * 2 * np.pi) + 1.5, r"$r^3$", fontsize=20)
plt.text(.95 * np.cos(x[45] * 2 * np.pi) + 1.5, .95 * np.sin(x[45] * 2 * np.pi) + 1.5, r"$r^2$", fontsize=20)
plt.text(.95 * np.cos(x[20] * 2 * np.pi) + 1.5, .95 * np.sin(x[20] * 2 * np.pi) + 1.5, r"$r$", fontsize=20)

plt.plot([1.33, 1.33], [1, 2], 'k-', alpha=.8)
plt.plot([1.67, 1.67], [1, 2], 'k-', alpha=.8)
plt.plot([1, 2], [1.33, 1.33], 'k-', alpha=.8)
plt.plot([1, 2], [1.67, 1.67], 'k-', alpha=.8)


plt.text(0.93, 1.03, "D", fontsize=20)
plt.text(2.03, 1.03, "C", fontsize=20)
plt.text(0.93, 1.9, "A", fontsize=20)
plt.text(2.03, 1.9, "B", fontsize=20)

plt.text(1.445, 1.438, "0", fontsize=50, alpha=.8)
plt.text(1.095, 1.425, "8", fontsize=50, alpha=.8)
plt.text(1.095, 1.765, "1", fontsize=50, alpha=.8)
plt.text(1.445, 1.765, "2", fontsize=50, alpha=.8)
plt.text(1.765, 1.765, "3", fontsize=50, alpha=.8)
plt.text(1.765, 1.425, "4", fontsize=50, alpha=.8)
plt.text(1.765, 1.095, "5", fontsize=50, alpha=.8)
plt.text(1.445, 1.095, "6", fontsize=50, alpha=.8)
plt.text(1.095, 1.095, "7", fontsize=50, alpha=.8)
plt.show()
```
